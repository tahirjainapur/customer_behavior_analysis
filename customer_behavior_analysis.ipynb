```python
def extract_duration(x):
    if pd.isna(x): return pd.NA
    x = x.lower()
    if "min" in x:
        return int(x.split("min")[0].strip())
    if "season" in x:
        return int(x.split("season")[0].strip())
    return pd.NA

df['duration_num'] = df['duration'].apply(extract_duration)
df['duration_num'].head()
```
